```python
"""
In this notebook we compare performance of the various preprocessing routines on the dataset
with 12k cases from Illinois with decision date after 1950.
"""

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
%matplotlib inline

results_header = '/Users/jhamer90811/Documents/Insight/legal_topic_modeling/validation_output'

preprocessor_headers = ['baseline', 'basic_stopwords', 'phrasing', 'phrasing_basic_stopwords',
                       'basic_stopwords_nouns', 'basic_stopwords_nouns_verbs']

dataset = 'cases_IL_after1950_12k'

```

```python
data = pd.DataFrame()
for prep in preprocessor_headers:
    new_data = pd.read_csv(os.path.join(results_header, prep + '_12k', dataset + '.csv'))
    new_data['preprocessor'] = prep
    data = data.append(new_data, ignore_index=True)
```

```python
# Use if stdev bars are desired

def plot_with_error_bars(df, x, y, y_err, group_by):
    groups = df.groupby(group_by)[[x, y, y_err]]

    fig, ax = plt.subplots()

    for k, v in groups:
        v.plot(label=k, x=x, y=y, ax=ax)
        plt.fill_between(v[x], v[y]-v[y_err], v[y]+v[y_err], alpha=0.1)

    ax.legend()
    ax.set_title(f'{y} vs. {x}; by {group_by}')

    plt.show()
```

```python
# Min citation distance
plot_with_error_bars(data, 'num_topics', 'min_cite_dist_mean', 'min_cite_dist_sd', 'name')
```

```python
# Average citation distance
plot_with_error_bars(data, 'num_topics', 'avg_cite_dist_mean', 'avg_cite_dist_sd', 'name')
```

```python
# Maximum citation distance
plot_with_error_bars(data, 'num_topics', 'max_cite_dist_mean', 'max_cite_dist_sd', 'name')
```
